```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread('./image/gate.jpeg')
imgp = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imgp)
```

```python
img.shape
```

```python
pre_in = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
pre_in.shape[0]
```

```python
result = np.zeros([1068,1750,3],dtype='uint8')
```

```python
for row in range(pre_in.shape[0]):
    for col in range(pre_in.shape[1]):
        result[row,col,2], result[row,col,1] = pre_in[row,col,2], pre_in[row,col,1]
        result[row,col,0] = pre_in[row,col,0]-20
```

```python
#result.shape
out = cv2.cvtColor(result, cv2.COLOR_HSV2BGR)
out = cv2.cvtColor(out, cv2.COLOR_BGR2GRAY)
out.shape
plt.imshow(out, cmap='gray')
```

```python
_, th = cv2.threshold(out,85,255,cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(th, cmap='gray')
```
